```python
import matplotlib.pyplot as plt
%matplotlib inline
import os
import numpy as np

from pyampp.gxbox.boxutils import read_b3d_h5
```

Load the b3d data with the dimensions of 64x64x64, 96x96x96, and 128x128x128, respectively. 

**Note: replace the datapath with your own path.**

```python
datapath = '/Users/fisher/myworkspace/'
b3dfile64 = os.path.join(datapath, 'b3d_data_20231014T160000_dim64x64x64.h5')
b3dfile96 = os.path.join(datapath, 'b3d_data_20231014T160000_dim96x96x96.h5')
b3dfile128 = os.path.join(datapath, 'b3d_data_20231014T160000_dim128x128x128.h5')
```

```python
b3dbox64 = read_b3d_h5(b3dfile64)
b3dbox96 = read_b3d_h5(b3dfile96)
b3dbox128 = read_b3d_h5(b3dfile128)
```

```python
b3dboxdict = {'64':b3dbox64, '96':b3dbox96, '128':b3dbox128}
def plot_b3d_z(zidx):
    fig, axs = plt.subplots(3, 4, figsize=(10, 7))
    for kidx, key in enumerate(['bx', 'by', 'bz']):
        dim, b3dbox = '64', b3dbox64
        ax = axs[kidx,0]
        ax.imshow(b3dbox['pot'][key][:, :,zidx])
        ax.set_title(f'z={zidx} {dim}x{dim}')
        ax.text(0.03,0.97, f'{key} pot', color='white', ha='left', va='top', transform=ax.transAxes,fontsize=12)        
        for icol, (dim, b3dbox) in enumerate(b3dboxdict.items()):
            zidx_adj = int(np.round(zidx/64*int(dim)))
            ax = axs[kidx,icol+1]
            ax.imshow(b3dbox['nlfff'][key][:, :,zidx_adj])
            ax.set_title(f'z={zidx_adj} {dim}x{dim}')
            ax.text(0.03,0.97, f'{key} nlfff', color='white', ha='left', va='top', transform=ax.transAxes,fontsize=12)
    fig.tight_layout()
    
```

```python
plot_b3d_z(0)
```
